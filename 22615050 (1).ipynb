```python
def euclid(a, b):
    if b == 0:
        return a
    else:
        return euclid(b, a % b)
print(euclid(12,24))
```

```python
def factorial(f):
    if f == 1:
        return 1
    else:
        return factorial(f - 1) * f
print(factorial(4))
```

```python
def factorial(f):
    fact = 1
    for i in range(2, f + 1):
        fact *= i
    return fact
print(factorial(4))
```

```python
def fibonacci_iterative(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    
    fib_sequence = [0, 1]
    for i in range(2, n):
        next_value = fib_sequence[-1] + fib_sequence[-2]
        fib_sequence.append(next_value)
    
    return fib_sequence

fib_sequence_iterative = fibonacci_iterative(30) 

print(f"반복문: {fib_sequence_iterative}")
```

```python
def fibonacci_memoized(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    memo[n] = fibonacci_memoized(n-1, memo) + fibonacci_memoized(n-2, memo)
    return memo[n]

fib_sequence_memoized = [fibonacci_memoized(i) for i in range(30)]

print(f"재귀적 함수: {fib_sequence_memoized}")
```

```python
def hanoi(n, from_peg, to_peg, by_peg):
    if n == 1:
        print(f"Move: {from_peg} -> {to_peg}")
    else:
        hanoi(n - 1, from_peg, by_peg, to_peg)
        print(f"Move: {from_peg} -> {to_peg}")
        hanoi(n - 1, by_peg, to_peg, from_peg)

hanoi(4, 'A', 'C', 'B')
```

```python
import time

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

import random
arr = [random.randint(1, 1000) for _ in range(100)]

start_time = time.time()
bubble_sort(arr)
end_time = time.time()
execution_time = end_time - start_time

print("배열:", arr[:])
print("실행 시간: {:.6f} 초".format(execution_time))
```

```python
import time
import random

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

arr = [random.randint(1, 1000) for _ in range(1000)]
start_time = time.time()
sorted_arr = quick_sort(arr)
end_time = time.time()
execution_time = end_time - start_time

print("배열:", sorted_arr[:])
print("실행 시간: {:.6f} 초".format(execution_time))
```

```python
import time
import random

def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    def merge(left, right):
        result = []
        i = j = 0

            
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                result.append(left[i])
                i += 1
            else:
                result.append(right[j])
                j += 1

        
        result.extend(left[i:])
        result.extend(right[j:])
        return result

   
    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])

   
    return merge(left_half, right_half)


arr = [random.randint(1, 1000) for _ in range(1000)]
start_time = time.time()
sorted_arr = merge_sort(arr)
end_time = time.time()
execution_time = end_time - start_time

print("배열:", sorted_arr[:])
print("실행 시간: {:.6f} 초".format(execution_time))
```
